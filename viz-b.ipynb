```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('diamonds')
df.head() 
```

```python
sns.relplot(x='clarity' ,y='price', data=df)
```

```python
modified_df = df.groupby('clarity', as_index=False).mean()
modified_df
```

```python
sns.relplot(x='clarity', y='price', data=modified_df)
```

```python
sns.barplot(x='clarity', y='price', data=modified_df)
```

```python
sns.lmplot(x='table', y='price', data=modified_df)
```

```python
sns.pointplot(x='table', y='price', data=modified_df)
```

```python
sns.set(rc={'figure.figsize':(14,8.27)})
sns.pointplot(x='clarity', y='price', data=modified_df)
```

### So this figures may appear like randomly selected, but they were chosen for a purpose! this is a different features made by the seaborn package. it is really useful tool if the data set are a real sample that have some requirement to be extracted from. it make your life easier!
